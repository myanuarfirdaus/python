```python
from tkinter import *
import math
class calculator:
    def clearall(self):
        e.delete(0,END)

    def action(self, num):
        e.insert(END,num)

    def equal(self):
        try:
            var1 = e.get()
            equiali=eval(var1)
            e.delete(0,END)
            e.insert(END,equiali)
        except SyntaxError or NameError:
            e.delete(0,END)
            e.insert(END, 'DATA ERRROR')

    def __init__(self, ventana):
        ventana.title("CALCULATOR")
        ventana.geometry("160x160")

ventana = Tk()
objek = calculator(ventana)
    
e = Entry(ventana)
e.grid(row=0,column=0,columnspan=4,pady=3)
e.focus_set()

#nomor tombol
Button(ventana, text="0", width=4, command=lambda:objek.action(0)).grid(column=1, row=4)
but1 = Button(ventana, text="1", width=4, command=lambda:objek.action(1))
but2 = Button(ventana, text="2", width=4, command=lambda:objek.action(2))
but3 = Button(ventana, text="3", width=4, command=lambda:objek.action(3))
but4 = Button(ventana, text="4", width=4, command=lambda:objek.action(4))
but5 = Button(ventana, text="5", width=4, command=lambda:objek.action(5))
but6 = Button(ventana, text="6", width=4, command=lambda:objek.action(6))
but7 = Button(ventana, text="7", width=4, command=lambda:objek.action(7))
but8 = Button(ventana, text="8", width=4, command=lambda:objek.action(8))
but9 = Button(ventana, text="9", width=4, command=lambda:objek.action(9))
#operasi
tambah = Button(ventana, text="+", width=4, command=lambda:objek.action('+'))
kurang = Button(ventana, text="-", width=4, command=lambda:objek.action('-'))
kali = Button(ventana, text="*", width=4, command=lambda:objek.action('*'))
bagi = Button(ventana, text="/", width=4, command=lambda:objek.action('/'))
samdeng = Button(ventana, text="=", width=4, command=objek.equal)
Button(ventana, text="AC", width=4, command=objek.clearall).grid(column=0, row=4)
#lokasi
but7.grid(column=0, row=1)
but8.grid(column=1, row=1)
but9.grid(column=2, row=1)
tambah.grid(column=3, row=1)

but4.grid(column=0, row=2)
but5.grid(column=1, row=2)
but6.grid(column=2, row=2)
kurang.grid(column=3, row=2)

but1.grid(column=0, row=3)
but2.grid(column=1, row=3)
but3.grid(column=2, row=3)
kali.grid(column=3, row=3)

bagi.grid(column=2, row=4)
samdeng.grid(column=3, row=4)


ventana.mainloop()
```
